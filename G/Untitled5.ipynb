```python
# ==========================
# IQテストゲーム
# 正解すると難易度アップ
# ==========================


levels = {
    1: [
        {
            "q": "2, 4, 6, 8, ?",
            "choices": ["9", "10", "12", "14"],
            "ans": 2,
            "exp": "2ずつ増えているので次は10です。"
        },
        {
            "q": "▲ ■ ▲ ■ ▲ ?",
            "choices": ["▲", "■", "●", "◆"],
            "ans": 2,
            "exp": "▲と■が交互に並んでいるので次は■です。"
        }
    ],

    2: [
        {
            "q": "3, 6, 12, 24, ?",
            "choices": ["36", "48", "50", "60"],
            "ans": 2,
            "exp": "毎回2倍になっています。24×2=48。"
        },
        {
            "q": "A, C, E, G, ?",
            "choices": ["H", "I", "J", "K"],
            "ans": 2,
            "exp": "アルファベットを1文字飛ばしで並べています。"
        }
    ],

    3: [
        {
            "q": "5人で5分で5個作る。100人なら100個作るのに何分？",
            "choices": ["5分", "20分", "100分", "1分"],
            "ans": 1,
            "exp": "1人が5分で1個作るので、100人なら100個を5分で作れます。"
        },
        {
            "q": "□→△→○→□→△→？",
            "choices": ["□", "△", "○", "☆"],
            "ans": 3,
            "exp": "□→△→○の繰り返しなので次は○です。"
        }
    ],

    4: [
        {
            "q": "1=1, 2=4, 3=9, 4=16, 5=?",
            "choices": ["20", "21", "25", "30"],
            "ans": 3,
            "exp": "数字を2乗しています。5²=25。"
        },
        {
            "q": "3, 6, 11, 18, 27, ?",
            "choices": ["34", "36", "38", "40"],
            "ans": 3,
            "exp": "増える数が3→5→7→9なので次は11。27+11=38。"
        }
    ],

    5: [
        {
            "q": "ある家族には父・母・娘がいる。娘には兄弟が2人いる。兄弟には兄弟が1人しかいない。子どもは全部で何人？",
            "choices": ["3", "4", "5", "6"],
            "ans": 1,
            "exp": "子どもは娘1人と兄2人で合計3人です。"
        },
        {
            "q": "1, 2, 6, 24, 120, ?",
            "choices": ["240", "360", "600", "720"],
            "ans": 4,
            "exp": "階乗の数列です。5!=120なので次は6!=720です。"
        }
    ]
}

iq = 100

print("=" * 40)
print("🧠 IQテストゲーム")
print("=" * 40)

for level in range(1, 6):

    print(f"\n★★★★★ レベル {level} ★★★★★")

    clear = True

    for quiz in levels[level]:

        print("\n" + quiz["q"])

        for i, c in enumerate(quiz["choices"], 1):
            print(f"{i}. {c}")

        while True:
            try:
                ans = int(input("番号を入力："))
                if 1 <= ans <= 4:
                    break
            except:
                pass
            print("1〜4を入力してください。")

        if ans == quiz["ans"]:
            print("⭕ 正解！")
            iq += 8
        else:
            print("❌ 不正解")
            print("正解は", quiz["ans"])
            iq -= 5
            clear = False
            break

    if clear:
        print(f"\n🎉 レベル{level}クリア！")
    else:
        print("\nゲームオーバー！")
        break



print("\n========================")
print("あなたの推定IQは", iq)
print("========================")

if iq < 90:
    print("判定：平均より少し低め")
elif iq < 110:
    print("判定：平均")
elif iq < 120:
    print("判定：平均以上")
elif iq < 130:
    print("判定：優秀")
elif iq < 140:
    print("判定：非常に優秀")
else:
    print("判定：天才クラス")





if ans == quiz["ans"]:
    print("\n⭕ 正解！")
    print("解説：" + quiz["exp"])
    iq += 8
elif "ans" in quiz and "choices" in quiz and quiz["ans"] - 1 < len(quiz["choices"]):
    print("\n❌ 不正解")
    print("正解：", quiz["choices"][quiz["ans"]-1])
    print("解説：" + quiz["exp"])
    iq -= 5
    clear = False
    # break  # 'break' is commented out as it would cause a SyntaxError outside a loop if this cell is run independently.
else:
    print("\n❌ 不正解")
    # Default message if 'ans' or 'choices' are missing or out of bounds
    print("解説：" + quiz["exp"] if "exp" in quiz else "解説がありません。")
    iq -= 5
    clear = False
    # break  # 'break' is commented out as it would cause a SyntaxError outside a loop if this cell is run independently.
```
